# Graph Theory Measure on Connectomes

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
features_path = '/workspaces/profCode/bctWrapper/features'
measure_list_path = 'bctWrapper/src/new/featureList.txt'
scanner_protocols = ('PNC_F', 'PNC_M', 'TOB_F', 'TOB_M')

measure_list = open(measure_list_path,'r').read().split('\n')
```

## Utilities functions

```python
def read_measure_output_file(measure_name):
    data = []
    for protocol in scanner_protocols:
        measure_path = os.path.join(features_path, protocol, measure_name + '.txt')
        arr = np.loadtxt(measure_path)
        if (len(arr.shape) > 1):
            arr = np.average(arr, axis=1)
        data.append(arr)
    return data
```

```python
def drawBoxPlot(data, title):
    fig, ax = plt.subplots()
    ax.boxplot(data)
    plt.xticks([1, 2, 3, 4], ['PNC Female', 'PNC Male', 'TOB Female', 'TOB Male'])
    plt.title(title)
    plt.show()
```

## Load measure output files and Visualize Box Plot

```python
for measure in measure_list:
    try: 
        data = read_measure_output_file(measure)
    except FileNotFoundError as e:
        print(f'File {measure}.txt not found')
    except ValueError as e:
        print(f'{e}')
    drawBoxPlot(data, measure)
        
```
